````python
PN_BASE_DIR = Path("../Pointnet_Pointnet2_pytorch/log/classification/")
PRED_FILE = PN_BASE_DIR / "msg_cls6_e40_bs8_pts4096_split7030_ds5fps/logs/pred_per_sample_7030.csv"

exp_name = PRED_FILE.parent.parent.name
print(exp_name)

df = pd.read_csv(PRED_FILE, header="infer")

display(df)
````

````python

df["max_conf"] = df[["bee","butterfly","dragonfly","wasp","bumblebee"]].max(axis=1)


with pd.option_context('display.float_format', '{:0.3f}'.format):
    display(df)
    display(df.describe(percentiles=[0.05,0.1,0.25,0.5,0.75]))
````

````python
#  5% of samples have max_conf >= x
conf_gt_m067_df = df[df.max_conf >= -0.670]
# 10% of samples have max_conf >= x
conf_gt_m055_df = df[df.max_conf >= -0.551]
# 25% of samples have max_conf >= x
conf_gt_m025_df = df[df.max_conf >= -0.254]
# 50% of samples have max_conf >= x
conf_gt_m006_df = df[df.max_conf >= -0.060]

print("all:", len(df.index))
print(len(conf_gt_m067_df.index))
print(len(conf_gt_m055_df.index))
print(len(conf_gt_m025_df.index))
print(len(conf_gt_m006_df.index))
````

````python

dfs = [df, conf_gt_m067_df, conf_gt_m055_df, conf_gt_m025_df, conf_gt_m006_df]
dfnames = ["df", "conf_gt_m067_df", "conf_gt_m055_df", "conf_gt_m025_df", "conf_gt_m006_df"]


print(f"Filter             Acc  #Sa.   %Sa.")
for dfi,dfname in zip(dfs, dfnames):
    acc = dfi[dfi.target_name == dfi.pred_name].shape[0] / dfi.shape[0]
    pct_samples = dfi.shape[0] / df.shape[0]
    print(f"{dfname: <16}", f"{acc: >.4f}", f"{dfi.shape[0]: >4} ", f"{pct_samples: >.3f}")
````
